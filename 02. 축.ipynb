# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' # 한글 폰트 설정, 맑은 고딕(window용)
# motplotlib.rcParams['font.family']='AppleGothic' # mac용
matplotlib.rcParams['font.size']=12 # 글자 크기
matplotlib.rcParams['axes.unicode_minus']=False #  한글 폰트 사용시 - 가 깨지는걸 방지
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺾은 선 그 래프',fontdict={'family':'HYGungSo-Bold','size':20}) # 개별 폰트도 설정가능하다.
```

```python
plt.plot(x,y)
plt.xlabel('X축') # x축 이름 설정
plt.ylabel('Y축')# y축 이름 설정
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red') # x축 이름 색 설정
plt.ylabel('Y축',color='#00aa00')# y축 이름 색 설정(RGB로 설정)
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red', loc='right') # x축 위치설정 , loc= left, center, right 가 들어갈 수 있다.
plt.ylabel('Y축',color='#00aa00', loc='top')# y축 위치설정, loc= top, center.bottom이 들어갈 수 있다.
```

```python
plt.plot(x,y)
plt.xticks([1,2,3]) # x축 틱 설정
plt.yticks([3,6,9,12]) # y축 틱 설정
plt.show()
```
