## COVID-19 백신 종류별 접종 인구
코로나 백신별 접종인구를 예시로 그래프를 그려보기

```python
days=[1, 2, 3] # 특정 '일'
az=[2, 4, 8] # (단위 : 만명) 1일~3일 까지 아스트라제네카 접종인구
pfizer=[5, 1, 3] # 화이자
moderna=[1, 2, 5] # 모더나

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='아스트로제네카')
plt.plot(days, pfizer, label='화이자',marker='o', ls='--')
plt.plot(days, moderna,label='모더나',marker='s', ls='-.')

plt.legend()
```

```python
plt.plot(days, az, label='아스트로제네카')
plt.plot(days, pfizer, label='화이자',marker='o', ls='--')
plt.plot(days, moderna,label='모더나',marker='s', ls='-.')

plt.legend(ncol=3) # ncol=3 을 하면 column을 3개만 표시할 수 있개해줌
```
