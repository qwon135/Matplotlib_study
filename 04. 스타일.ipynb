# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=12
matplotlib.rcParams['axes.unicode_minus']=False 
```

```python
x=[1,2,3]
y=[2,4,8]
```

### 가장 기본 그래프

```python
plt.plot(x,y)
```

### 선 굵기 조절

```python
plt.plot(x,y,linewidth=5)
```

### 데이터 있는 부분에 점 찍기

```python
plt.plot(x,y,marker='o')
```

### 데이터 있는 부분에 점만 찍기

```python
plt.plot(x,y,marker='o',linestyle='None')
```

### 데이터 있는 부분을 화살표로

```python
plt.plot(x,y,marker='v', markersize=10)
```

### 데이터 있는 부분을 X표시

```python
plt.plot(x,y,marker='X', markersize=10)
```

marker의 종류 :  
https://matplotlib.org/stable/api/markers_api.html

### Marker 테두리 색 바꾸기

```python
plt.plot(x,y,marker='o', markersize=20, markeredgecolor='red')
```

### Marker 색 바꾸기

```python
plt.plot(x,y,marker='o', markersize=20, markeredgecolor='red',markerfacecolor='yellow')
```

## 선 스타일

### 선 점선으로 표현

```python
plt.plot(x,y,linestyle=':')
```

### 선 점선(두꺼운) 으로 표현

```python
plt.plot(x,y,linestyle='--')
```

### 선을 점, 점선 으로 표현

```python
plt.plot(x,y,linestyle='-.')
```

line style 종류 :  
https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

## 포맷

선 색, marker종류, line style 등을 간편하게 입력하는 방법이 있다.

```python
plt.plot(x,y,'ro--') # (color,marker,linestyle) = ( red, 동그라미marker , 점선)
```

```python
plt.plot(x,y,'bv:') # 파란색,화살표 marker, 점선(클론)
```

```python
plt.plot(x,y,'go') # 초록색, 동그라미 marker , 선의 정보가 없어 선은 안나타난다.
```

## 축약어

너무 긴 단어들의 경우  
marker face color = mfc  
marker size=ms  
markeredgecolor=mec  
line style=ls  
line width=lw   
의 축약어를 사용할 수 있다.

축약어 정보 :  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(x,y,marker='o',mfc='r',ms='10',mec='b', ls=':')
```

## 투명도

```python
plt.plot(x,y,lw=10,marker='o',ms=20, alpha=0.3) # alpha 는 0~1의 범위를 가진다.
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x,y)
```

### dpi 설정
inchi당 도트 수(=해상도)

```python
plt.figure(figsize=(10, 5),dpi=100)
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
