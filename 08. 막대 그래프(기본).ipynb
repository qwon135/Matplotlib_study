# 8. 막대 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=12
matplotlib.rcParams['axes.unicode_minus']=False 
```

#### 막대 그래프 그리기

```python
labels=['강백호', '서태웅', '정대만'] # 이름
values=[190, 187, 184] # 키

plt.bar(labels,values)
```

#### 막대 색 변경

```python
plt.bar(labels, values, color='r')# 색 변경
```

#### 막대마다 다른 색 설정

```python
labels=['강백호', '서태웅', '정대만'] # 이름
values=[190, 187, 184] # 키
colors=['r','g','b'] # index마다 다른 색 

plt.bar(labels,values,color=colors, alpha=0.5) # 투명도도 설정가능
```

#### plt.xlim/ylim 사용
'키'의 값이 비슷해 그래프가 잘 구별안되는 문제 해결

```python
plt.bar(labels,values)
plt.ylim(175,195) # 175~195까지만 보여준다.
```

#### bar 두께 조절

```python
plt.bar(labels, values, width=0.5)
```

#### 글씨 기울이기 (겹치는 경우 대비)

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # x축 이름 각도 45도로 설정
plt.yticks(rotation=45) # y축 이름 각도 45도로 설정
plt.show()
```

#### ticks 이름 바꾸기

```python
labels=['강백호', '서태웅', '정대만'] # 이름
values=[190, 187, 184] # 키
ticks=['1번 학생','2번 학생','3번 학생']

plt.bar(labels, values,width=0.5)
plt.xticks(labels, ticks)
plt.show()
```
