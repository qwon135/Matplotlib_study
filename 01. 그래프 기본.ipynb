# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x=[1,2,3]
y=[2,4,8]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show() # show() 를하면 위의 불필요한 데이터 없이 그래프만 출력
```

## Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은 선 그래프') # 한글을 쓰려면 추가 환결설정을 해야한다.
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' # 한글 폰트 설정, 맑은 고딕(window용)
# motplotlib.rcParams['font.family']='AppleGothic' # mac용
matplotlib.rcParams['font.size']=12 # 글자 크기
matplotlib.rcParams['axes.unicode_minus']=False #  한글 폰트 사용시 - 가 깨지는걸 방지
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 리스트 형태로 이름만 출력
```
