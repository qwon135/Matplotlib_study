# 9. 막대 그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=12
matplotlib.rcParams['axes.unicode_minus']=False 
```

### 옆으로 누운 막대그래프

```python
labels=['강백호', '서태웅', '정대만'] # 이름
values=[190, 187, 184] # 키

plt.barh(labels,values)# bar대신 barh를 사용
plt.xlim(175,195) # 이번엔 ylim이 아닌 xlim으로 설정
```

### 막대에 모양 넣기
hatch 모양 관련 정보 :  
https://matplotlib.org/devdocs/gallery/shapes_and_collections/hatch_style_reference.html

```python
bar=plt.bar(labels, values,width=0.5)
bar[0].set_hatch('/') # 해당 막대에 ///모양이 생긴다.
bar[1].set_hatch('x')# xxxx
bar[2].set_hatch('*') # ****
```

### 막대에 글자 넣기
 plt.text(x,y,text) 에서 x는 idx, y는 rect에서 세로값을 가져온다, text에는 키를 넣을 것이므로 values[idx]이다.

```python
bar=plt.bar(labels, values,width=0.5)
plt.ylim(175,195)

for idx,rect in enumerate(bar):
    plt.text(idx,rect.get_height()+0.5 ,values[idx],ha='center', color='r', size=15) 
```
