#### 격자 넣기

```python
plt.plot(df['지원번호'],df['영어'])
plt.plot(df['지원번호'],df['수학'])

plt.grid() # grid는 격자를 넣어준다.
```

```python
plt.plot(df['지원번호'],df['영어'])
plt.plot(df['지원번호'],df['수학'])

plt.grid(axis='y',color='r', alpha=0.5, ls='--', lw=2) # y축에만 격자 넣기, 색 넣기, 투명도, linestyle설정 모두 가능
```
