# 6. 텍스트
데이터가 많을 땐 각 데이터마다 어떤 데이터인지 글을 달고 싶을 때

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=12
matplotlib.rcParams['axes.unicode_minus']=False 
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.figure(dpi=100)
plt.plot(x,y,marker='o')

for idx,txt in enumerate(y): # plt(x,y,text) 를하면 해당 좌표에 글자를 넣어준다. enumerate를 사용해 y값을 넣어준다.
    plt.text(x[idx],y[idx]+0.3,txt, ha='center', color='b') # 보기편하게 0.3을 더해 글자를 약간 올려준다. / ha='center'를 하면 글자를 중간 정렬해줌
```
